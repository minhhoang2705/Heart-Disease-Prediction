```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns

dataset_dir = "../Heart-Disease-Prediction/data/cleveland.csv"

df = pd.read_csv(dataset_dir, header=None)
df.columns = ['age', 'sex', 'cp', 'trestbps', 'chol',
              'fbs', 'restecg', 'thalach', 'exang',
              'oldpeak', 'slope', 'ca', 'thal', 'target']

df['target'] = df['target'].map({0: 0, 1: 1, 2: 1, 3: 1, 4: 1})
df['thal'] = df['thal'].fillna(df['thal'].mean())
df['ca'] = df['ca'].fillna(df['ca'].mean())

df.to_csv('../Heart-Disease-Prediction/data/cleveland_processed.csv', index=False)
# distribution of target vs age
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='target', multiple='stack', bins=20, kde=True)
plt.title('Distribution of Age vs Target')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(labels=['No Disease', 'Disease'], title='Target')


plt.show()
```

```python
# barplot of age vs sex with hue = target

plt.figure(figsize=(12, 6))
sns.barplot(x='sex', y='age', hue='target', data=df, palette=['skyblue', 'orange'])

plt.title('Variation of Age for each target class', fontsize=14)
plt.xlabel('Sex', fontsize=12)
plt.ylabel('Age', fontsize=12)

# Customize x-axis labels
plt.xticks([0, 1], ['Female', 'Male'])

# Customize legend
plt.legend(title='Target', labels=['No Disease', 'Disease'])

# Remove top and right spines
sns.despine()

# Adjust y-axis to start from 0
plt.ylim(0, df['age'].max() * 1.1)  # 10% margin on top

plt.tight_layout()
```
